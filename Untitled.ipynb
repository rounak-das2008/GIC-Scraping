```python
yearbook_url = "https://www.gicouncil.in/yearbook/2019-20/index"
```

```python
def collect_annexure_link(index_url):
    r = requests.get(index_url, verify=False)
    soup = BeautifulSoup(r.content, "html.parser")
    sleep(3)

    def get_annexure_links(tag):
        return(tag.name == 'a' and 
               'href' in tag.attrs and
               'https://www.gicouncil.in/yearbook/2019-20/annexures/' in tag['href'].lower())
    results = soup.find_all(get_annexure_links)

    results.pop(0)
    for result in results:
        print(result['href'])
    return results
```

```python
def collect_dataframes(df_url, folder_name):
    file_name = df_url.split("/")[-2]
    dfs = pd.read_html(df_url)
    for df in dfs:
        df.to_csv(f"{folder_name}/{file_name}.csv", index=False, header=None, mode='a')
        with open(f"{folder_name}/{file_name}.csv", 'a') as f:
            f.write('\n')
    sleep(3)
    return file_name
    
```

```python
annexure_links = collect_annexure_link(yearbook_url)
folder_name = yearbook_url.split("/")[-2]
os.mkdir(folder_name)
print("In Main --------")
for annexure_link_a in annexure_links:
    print(annexure_link_a)
    table_link = annexure_link_a['href']
    file_name = collect_dataframes(table_link, folder_name)
    print(" ====== ", file_name, ": Done =========")
print(" ====== ", folder_name, ": Done =========")
```
